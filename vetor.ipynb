```python
x = 3
v = [None]*x

for i in range(x):
    v[i] = float(input())

for i in range( 0,x-1):
    for j in range(i+1,x):
        if v[i] > v[j]:
            aux = v[i]
            v[i] = v[j]
            v[j] = aux
print(v)
    
```

```python
x = 3
v = [None]*x

for i in range(x):
    v[i] = float(input())

m = (x // 2)
soma = v[0]+v[x-1]
ini = 0
fim = x-1
for i in range (0, m):
    if (v[i] + v[x-i-1]  > soma):
        soma = v[i]+v[x-i-1]
        ini = i
        fim = x-i-1
print ("inicio:", ini, "fim:", fim, "soma:", soma)
```
